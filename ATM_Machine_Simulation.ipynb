ATM Machine Simulation

We create a small ATM program where user insert atm pin and od check balance and withdrawal draw the ammount.

```python
# Step 1: Initial balance and pin
pin = 1234
balance = 10000

# Step 2: Function to check pin
def check_pin(input_pin):
    if input_pin == pin:
        return True
    else:
        return False

# Step 3: Show menu
def show_menu():
    print("\nWelcome to the ATM")
    print("1. Check balance")
    print("2. Deposit money")
    print("3. Withdraw money")
    print("4. Exit")
    choice = int(input("Enter your choice: "))
    return choice

# Step 4: Check balance
def check_balance():
    print("Your balance is: ‚Çπ", balance)

# Step 5: Deposit money
def deposit_money():
    global balance
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print("Deposit successful. New balance: ‚Çπ", balance)

# Step 6: Withdraw money
def withdraw_money():
    global balance
    amount = float(input("Enter amount to withdraw: "))
    if amount > balance:
        print("‚ùå Insufficient balance")
    else:
        balance -= amount
        print("Withdrawal successful. New balance: ‚Çπ", balance)

# Step 7: Main function
def main():
    input_pin = int(input("Enter your PIN: "))
    
    if not check_pin(input_pin):  # <<--- yahan NOT ka use hua hai
        print("‚ùå Incorrect PIN. Try again.")
        return

    print("‚úÖ PIN is correct")

    while True:
        choice = show_menu()
        if choice == 1:
            check_balance()
        elif choice == 2:
            deposit_money()
        elif choice == 3:
            withdraw_money()
        elif choice == 4:
            print("üëã Thank you for using the ATM. Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
main()
```
